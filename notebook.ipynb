```python
import duckdb
import pandas as pd
```

```python
df = pd.read_csv('https://covid.ourworldindata.org/data/owid-covid-data.csv')
```

```python
# 1.2 Get the basic info
df.info()
```

```python
# 1.3 Sample 10 observations
df.sample(10)
```

```python
# 1.4 Let's use a SQL query to select the first 10 rows in the table
duckdb.query("SELECT * FROM df LIMIT 10;")
```

```python
# 1.5 Show the results as a data frame
duckdb.query("SELECT * FROM df LIMIT 10;").to_df()
```

```python
# 1.6 Retrieve the total number of deaths in Peru, until 31-12-2022

duckdb.query("SELECT iso_code, location, new_deaths FROM df WHERE location='Peru' OR location='Guatemala' AND date<='2022-12-31' LIMIT 10;").to_df()
```

```python
# 1.6 Retrieve the total number of deaths in Peru, until 31-12-2022

duckdb.query("SELECT SUM(new_deaths) AS total FROM df WHERE location='Peru' OR location='Guatemala' AND date<='2022-12-31' LIMIT 10;").to_df()
```

```python
# 1.7 Retrieve the total number of deaths in North and South America, until 31-12-2022
duckdb.query("SELECT DISTINCT(continent) FROM df;").to_df()
```

```python
duckdb.query("SELECT DISTINCT(location) FROM df;").to_df()
```

```python
duckdb.query("SELECT SUM(new_deaths) AS total FROM df WHERE continent IN('North America','South America') AND date<='2022-12-31' LIMIT 10;").to_df()
```

```python
duckdb.query("SELECT DISTINCT(location) FROM df WHERE continent='North America';").to_df()
```

```python
# 1.8 Summarise the total number of deaths by continent
duckdb.query("SELECT continent, SUM(new_deaths) AS total FROM df GROUP BY continent;").to_df()
```

```python
# 1.9 What was the total number of deaths by continent, considering America as 
# North America and South America? Present the results ordered by total in 
# descending order

duckdb.query("SELECT CASE WHEN continent ='North America' THEN 'America' WHEN continent ='South America' THEN 'America' ELSE continent END AS the_continent, SUM(new_deaths) AS total FROM df WHERE the_Continent IS NOT NULL AND  date<='2022-12-31' GROUP BY the_continent ORDER BY total desc;").to_df()
```
